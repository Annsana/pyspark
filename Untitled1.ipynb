```python
df_pyspark_tran = spark.read.csv('Tran_creditcard.csv',header=True,inferSchema=True)
```

```python
df_pyspark_tran.show(10)
```

```python
df_pyspark_tran.printSchema()
```

```python
###Partitioning on Tran_creditcard
from pyspark.sql.functions import year, month, col

df_part = df_pyspark_tran.withColumn("year",year(col("tran_date"))).withColumn("month",month(col("tran_date")))
```

```python
df_part.write.partitionBy("year", "month").mode("overwrite").saveAsTable("tran_creditcardpartiitoned")
```
